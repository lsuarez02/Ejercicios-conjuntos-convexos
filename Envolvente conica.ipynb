A continuación se desarrollará una idea para encontrar la envolvente cónica de un conjunto de puntos. Esta se dividirá en dos partes.

##### 1. Puntos en $\mathbb{R}^2$

Si los puntos a considerar se encuentran en $\mathbb{R}^2$ y satisfacen que todos tienen su primera componente con el mismo signo, se procede así:

using PlotlyJS

A=[x1 x2; y1 y2; z1 z2,...,k1];  # se ingresan los puntos en una matriz

n=length(A)/2;                   # se indica la cantidad de puntos

for i=1:n

    f[i]=(A[i,2]/A[i,1]*x);      # se crea el rayo de (0,0) a cada punto
    
end

for i=1:n-1

    plot(x,f[i],fillrange=f[i+1]); # se grafican los puntos entre cada dos rayos.
    
end

Algunos ejemplos en donde se aplica el algoritmo son

```julia
using PlotlyJS
A=[1 4; 5 2];
f1(x)=(A[1,2]/A[1,1])*x;
f2(x)=(A[2,2]/A[2,1])*x;
plot(x, f1, fillrange=f2, title="Envolvente cónica de dos puntos")
```

```julia
using PlotlyJS
using Random
A=[1 4;2 -3];
f1(x)=(A[1,2]/A[1,1])*x;
f2(x)=(A[2,2]/A[2,1])*x;
plot(x, f2, fillrange=f1)
```

Al juntar los tres puntos de los anteriores ejemplos, tenemos

```julia
using PlotlyJS
A=[1 4; 5 2; 2 -3];
f1(x)=(A[1,2]/A[1,1])*x;
f2(x)=(A[2,2]/A[2,1])*x;
f3(x)=(A[3,2]/A[3,1])*x;
plot(x, f1, fillrange=f2, )
plot!(x, f2, fillrange=f3)
plot!(x,f3)#, fillrange=f1)
```

Sin embargo, el procedimiento anterior no se puede realizar en dimensiones mayores, por lo que haremos algo diferente. Sabemos que la envolvente cónica de un conjunto es una combinación $Ax$, donde $A$ es la matriz que contiene los puntos y $x$ son los puntos en $\mathbb{R}^n$ que satisfacen $x_i\geq0$ para todo $i$. Usaremos esta definición para verificar si un punto pertenece o no a la envolvente cónica de un conjunto. Así, por ejemplo

```julia
A=[1 3 5 0 1; 3 4 2 1 0; 1 0 1 0 1];
b=[1; 1 ;2];
s=A\b
```

```julia
if all(>=(0),s)  
    print("b pertenece a la envolvente cónica")
else
    print("b no pertenece a la envolvente cónica")
end
```
